```python
subtractive_lu = {'IV':'IIII', 'IX':'VIIII', 'XL':'XXXX', 'XC':'LXXXX', 'CD':'CCCC', 'CM':'DCCCC'}
def replace_subtractives(rn):
    res = rn
    for old,new in subtractive_lu.items():
        res = res.replace(old, new)
    return res
```

```python
roman_lu = {'M':1000, 'D':500, 'C':100, 'L':50, 'X':10, 'V':5, 'I':1}

def from_roman(rn):
    nums = [roman_lu[c] for c in replace_subtractives(rn)]
    return sum(nums)
```

```python
def to_roman(num):
    rn = 'M'*(num//1000)
    rem = num % 1000
    if rem >= 900:
        rn += 'CM'
        rem -= 900
    elif rem >= 500:
        rn += 'D'
        rem -= 500
    elif rem >= 400:
        rn += 'CD'
        rem -= 400
    rn += 'C'*(rem//100)
    rem = rem % 100
    if rem >= 90:
        rn += 'XC'
        rem -= 90
    elif rem >= 50:
        rn += 'L'
        rem -= 50
    elif rem >= 40:
        rn += 'XL'
        rem -= 40
    rn += 'X'*(rem//10)
    rem = rem % 10
    if rem == 9:
        return rn + 'IX'
    elif rem >= 5:
        rn += 'V'
        rem -= 5
    elif rem == 4:
        return rn + 'IV'
    return rn + 'I'*rem
```
